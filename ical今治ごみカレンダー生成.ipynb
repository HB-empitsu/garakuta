# カレンダー生成

```python
import pandas as pd
```

```python
# 日付
start_date = f"{year}-04-01"
end_date = f"{year + 1}-03-31"
```

```python
# 正月三が日
dates_to_remove = [f"{year + 1}-01-01", f"{year + 1}-01-02", f"{year + 1}-01-03"]
```

```python
# 年度の日付生成
date_index = pd.date_range(start=start_date, end=end_date, freq="D")
```

```python
df_date = pd.DataFrame(index=date_index)
```

```python
# 正月三が日を削除
df_date = df_date.drop(dates_to_remove, axis=0, errors="ignore")
```

```python
# 曜日名辞書生成
weekday = {i: v for i, v in enumerate("月火水木金土日")}
```

```python
# 週目
df_date["week_number"] = (df_date.index.day - 1) // 7 + 1

# 曜日
df_date["weekday"] = df_date.index.weekday
df_date["week_name"] = df_date["weekday"].apply(lambda x: weekday[x])
```

```python
# 曜日パターン生成
days = {}

for 曜日 in "月火水木金土日":

    days[f"第１・３{曜日}"] = df_date[df_date["week_number"].isin([1, 3]) & (df_date["week_name"] == 曜日)].index
    days[f"第２・４{曜日}"] = df_date[df_date["week_number"].isin([2, 4]) & (df_date["week_name"] == 曜日)].index

    days[曜日] = df_date[df_date["week_name"] == 曜日].index
```

# エリア

```python
import io
```

```python
df_area = pd.read_csv(io.StringIO(area_csv), index_col=0)
df_area
```

```python
# エリア抽出
d = df_area.loc[area].to_dict()
```

```python
# ごみ種類ごとの日付生成
data = [pd.Series("可燃", index=days[i]) for i in d["可燃"].split("・")]

data.append(pd.Series("不燃", index=days[d["不燃"]]))
data.append(pd.Series("プラ", index=days[d["プラ"]]))
data.append(pd.Series("資源", index=days[d["資源"]]))
```

```python
# ごみ種類別をひとつに結合
sr = pd.concat(data)
sr
```

```python
# ごみの種類ごとに横に並べる
df0 = sr.reset_index(name="value").pivot(index="index", columns="value", values="value").reindex(columns=["可燃", "不燃", "プラ", "資源"])
df0
```

```python
# 同日はごみ種類を結合
df1 = df0.apply(lambda s: "・".join(s.dropna()), axis=1).to_frame(name="ごみ")
```

```python
# タイムゾーンを日本に設定
df1["開始"] = (df1.index + pd.Timedelta("8:30:00")).tz_localize("Asia/Tokyo")
df1["終了"] = (df1.index + pd.Timedelta("9:00:00")).tz_localize("Asia/Tokyo")
```

```python
df1
```

# icalファイル生成

```python
from ics import Calendar, Event
```

```python
import pathlib
```

```python
c = Calendar()
c.serializecreator = "えんぴつ"


for i, row in df1.iterrows():
    event = Event()
    event.name = row["ごみ"]
    event.begin = row["開始"]
    event.end = row["終了"]
    # event.description = row["Description"]
    c.events.add(event)
```

```python
fn = pathlib.Path(f"{area}.ics")
```

```python
with open(fn, "w", encoding="utf-8") as f:
    f.write(c.serialize())
```
