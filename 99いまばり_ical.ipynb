```python
import pandas as pd
```

```python
# CSVファイル読み込み
df0 = pd.read_csv("https://imabari.jpn.org/99imabari/latest.csv", index_col=0, parse_dates=["date"])
```

```python
# 同日2回を分割
df0["time"] = df0["time"].str.split(" / ")
```

```python
# 分割したtimeを展開
df1 = df0.explode("time").reset_index(drop=True)
```

```python
df1
```

```python
# 開始時間と終了時間に分解
df1[["開始時間", "終了時間"]] = df1["time"].str.split("～", expand=True)
```

```python
# timedeltaに変換して翌日の場合は+1日
def timeconv(se):
    df = se.str.extract("(?P<翌日>翌日)?(?P<時間>\d{2}:\d{2})", expand=True)
    df["time"] = pd.to_timedelta(df["時間"] + ":00")
    df["time"].mask(df["翌日"] == "翌日", df["time"] + pd.Timedelta(days=1), inplace=True)
    return df["time"]
```

```python
# 日付に時間を付与、タイムゾーンを日本に設定
df1["start"] = (df1["date"] + timeconv(df1["開始時間"])).dt.tz_localize("Asia/Tokyo")
df1["end"] = (df1["date"] + timeconv(df1["終了時間"])).dt.tz_localize("Asia/Tokyo")
```

```python
# タイトル用に診療科目と病院名を結合
df1["title"] = df1.apply(lambda x: f'【{x["type"]}】{x["name"]}', axis=1)
```

```python
from ics import Calendar, Event
```

```python
c = Calendar()
c.serializecreator = "99いまばり"

for i, row in df1.iterrows():
    event = Event()
    event.name = row["title"]
    event.begin = row["start"]
    event.end = row["end"]
    event.location = row["address"]
    c.events.add(event)
```

```python
with open("99imabari.ics", "w", encoding="utf-8") as f:
    f.write(c.serialize())
```
